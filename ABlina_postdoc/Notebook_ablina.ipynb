```python
list_lipids=['POPC','POPE','POPS','CHOL','DPSM','GM1','GM3']
df_grouped_norm = df_mgrouped.copy()
for mem in (df_grouped_norm.index):
    for lip in list_lipids:
        contacts = "Contacts_m"+str(lip); composition = "#_"+str(lip)
        if df_grouped_norm.loc[mem, composition] == 0:
            relation = 0
        else:
            perc_contacts = df_grouped_norm.loc[mem, contacts] / df_grouped_norm.loc[mem, 'Contacts_mtotal']
            perc_composition = df_grouped_norm.loc[mem, composition] / df_grouped_norm.loc[mem, '#_total_lipids']
            relation = perc_contacts / perc_composition
        df_grouped_norm.loc[mem, contacts] = relation; df_grouped_norm.loc[mem, composition] = perc_composition
#df_grouped_norm
```

```python
df_grouped_norm.rename(columns={'Contacts_mPOPC': 'POPC', 'Contacts_mPOPE':'POPE','Contacts_mPOPS': 'POPS','Contacts_mCHOL': 'CHOL','Contacts_mDPSM':'DPSM','Contacts_mGM1':'GM1','Contacts_mGM3':'GM3'}, inplace=True)
```

```python
fig=plt.figure(figsize=(12, 6))
ax = fig.gca()
df_grouped_norm[['POPC','POPE','POPS','CHOL','DPSM','GM1']].plot(kind="barh", ax=ax)
plt.ylabel(None)
plt.title('Normalized Aβ-lipid contacs', fontsize=14, fontweight='bold')
plt.xlabel('Normalized # of contacts', fontsize=14)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), frameon=False)
plt.tight_layout()
plt.savefig('FIG8A_normalized-numb-contacts.png')
plt.show()
```

```python
orden_mem=['PC-GM3 8:1', 'PC-GM1 8:1', 'PC-GM3 20:1', 'PC-GM1 20:1', 'PC-PE-PS-CHOL-DPSM-GM3 8:5:1:4:2:4', 'PC-PE-PS-CHOL-DPSM-GM1 8:5:1:4:2:4', 'PC-PE-PS-CHOL-DPSM-GM3 8:5:1:4:2:2', 'PC-PE-PS-CHOL-DPSM-GM1 8:5:1:4:2:2', 'PC-PE-PS-CHOL-DPSM-GM3 8:5:1:4:2:1', 'PC-PE-PS-CHOL-DPSM-GM1 8:5:1:4:2:1']
```

```python
fig=plt.figure(figsize=(12, 6))
ax = fig.gca()
df_contacts_grouped = df_contacts.groupby(['membrane'])[['Contacts_glc','Contacts_gal-int','Contacts_nana','Contacts_galnac','Contacts_gal-ext','Contacts_tails','Contacts_total']].mean()
df_contacts_grouped.rename(columns={'Contacts_glc': 'β-D-Glc', 'Contacts_gal-int':'β-D-Gal (1)','Contacts_nana': 'NANA','Contacts_galnac': 'β-D-GalNAc','Contacts_gal-ext':'β-D-Gal (2)','Contacts_tails':'Tails'}, inplace=True)

para_ordenar=df_contacts_grouped.copy()
para_ordenar.reset_index(inplace=True)
para_ordenar["membrane"]=pd.Categorical(para_ordenar["membrane"], categories=orden_mem, ordered=True)
para_plotear=para_ordenar.set_index(para_ordenar["membrane"]).loc[orden_mem]

ax = para_plotear[['β-D-Glc','β-D-Gal (1)','NANA','β-D-GalNAc','β-D-Gal (2)','Tails']].plot(kind="barh", ax=ax)
ax.set_title('Aβs-sugar units contacts', fontsize=14, fontweight='bold')
ax.set_ylabel(None)
ax.set_xlabel('# of contacts', fontsize=14)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), frameon=False, title="Sugar units")
plt.tight_layout()
plt.savefig('FIG12_contacts-sugars.png')
```
